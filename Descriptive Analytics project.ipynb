# Histogram

```python
plt.rcParams['figure.figsize']=(18,8)

plt.subplot(1,2,1)
sns.histplot(df['total_bill'],kde=True)

plt.subplot(1,2,2)
sns.histplot(df['tip'],kde=True)
             
plt.suptitle('Univariate Analysis on Numerical Columns')
plt.show()
```

# Box Plot

```python
plt.subplot(1,2,1)
sns.boxplot(df['total_bill'])

plt.subplot(1,2,2)
sns.boxplot(df['tip'])

plt.suptitle('Outliers in the data')
plt.show()

```

# Scatter Plot

```python

sns.scatterplot(x='total_bill',y='tip',data=df,hue='sex',style='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,size='sex',col='size',col_wrap=2)
plt.show()
```

# Line Plot

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='line')
plt.show()
```

# Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(18,8)

plt.subplot(3,2,1)
sns.countplot(x=df['sex'])

plt.subplot(3,2,2)
sns.countplot(x=df['smoker'])

plt.subplot(3,2,3)
sns.countplot(x=df['day'])

plt.subplot(3,2,4)
sns.countplot(x=df['time'])

plt.subplot(3,2,5)
sns.countplot(x=df['size'])

plt.suptitle('Univariate Analysis on Discrete Columns')
plt.show()
```
